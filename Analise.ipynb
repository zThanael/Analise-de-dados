```python
import os
import pandas as pd
import sqlite3
import matplotlib as plt
```

```python
banco = sqlite3.connect('pesquisa_satisfacao.db')
```

### Criar a consulta SQL para trazer os dados que quero do banco

```python
cur = banco.cursor()
sql = '''SELECT p.enunciado, r.resposta 
FROM perguntas p
INNER JOIN respostas r
    ON p.id = r.id_pergunta
'''
cur.execute(sql)
query = cur.fetchall()
#query é uma lista com todas as respostas
```

### Criar nosso DataFrame

```python
df = pd.read_sql(sql,banco)
```

```python
df #O DataFrame com as perguntas e as respostas
```

```python
def criar_lista(df,coluna):
    lista = []
    for dado in df[coluna]:
        lista.append(dado)
    return list(set(lista))

def criar_filtro(df,lista,coluna):
    dados = {}
    for dado in lista:
        mask = (df[coluna] == dado)
        dados[dado] = (df[mask])
    return dados
```

```python
perguntas = criar_lista(df,'enunciado') #Salvar em uma lista todas as perguntas.
```

```python
perguntas
```

```python
dados_perguntas = criar_filtro(df,perguntas,'enunciado') #Criar o filtro de todas as perguntas.
```
